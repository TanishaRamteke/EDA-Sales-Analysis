```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("dataset.csv", encoding='latin1') 
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
high_sales_df =df[df["Sales"] > 1000]
high_sales_df
```

```python
region_sales = high_sales_df.groupby("Region")["Sales"].sum()
region_sales
```

```python
df['Profit_Status'] = df['Profit'].apply(
    lambda x: 'Positive' if x > 0 else 'Loss'
)

df[["Profit","Profit_Status"]].head()
```

```python
df['Order_Size'] = df['Quantity'].apply(
    lambda x:'Large' if x>5 else 'Small'
)
df['Order_Size'].head(10)
```

```python
top_region = region_sales.sort_values(ascending=False).head()
top_region
```

```python
high_profit_sales = df[(df["Sales"]>1000)& (df["Profit"] > 0)]

high_profit_sales.groupby("Region")["Sales"].sum().sort_values(ascending=False).head()

```

```python
fig, ax = plt.subplots(figsize=(8,5))

avg_profit = df.groupby('Category')['Profit'].mean().reset_index()

sns.barplot(x='Category', y='Profit', data=avg_profit, ax=ax)

ax.set_title('Average Profit by Category')
ax.set_xlabel('Category')
ax.set_ylabel('Average Profit')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.boxplot(x='Region',y='Sales', data=df)
ax.set_title('Sales Distribution by Region')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))

corr = df[['Sales', 'Profit', 'Quantity']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', ax=ax)
ax.set_title('Correlation Heatmap')

plt.show()

```
